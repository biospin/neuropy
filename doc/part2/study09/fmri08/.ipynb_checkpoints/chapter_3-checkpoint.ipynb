(2) 기본 그래프 그리기 : barchart, horizontal barchart, boxplot, scatterplot

```python
from matplotlib.pyplot import *

# some simple data
x = [1,2,3,4]
y = [5,4,3,2]

# create new figure
figure()

# divide subplots into 2 x 3 grid
# and select #1
subplot(231)
plot(x, y)

# select #2
subplot(232)
bar(x, y)

# horizontal bar-charts
subplot(233)
barh(x, y)

# create stacked bar charts
subplot(234)
bar(x, y)

# we need more data for stacked bar charts
y1 = [7,8,5,3]
bar(x, y1, bottom=y, color = 'r')

# box plot
subplot(235)
boxplot(x)

# scatter plot
subplot(236)
scatter(x,y)

show()
```

- bar()에 bottom = y로 주면 y를 아래에 놓고 위에 y1을 쌓아서  stacked barchart를 그린다.  
- boxplot()은 이름과 같이 boxplot을, scatter()는 scatterplot을 그린다.

## 3.2 Drawing a simple sine and cosine plot

(1) 기본적인 cos, sin 그래프 그리기

```python
import matplotlib.pyplot as pl
import numpy as np

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

y = np.cos(x)
y1 = np.sin(x)

pl.plot(x,y)
pl.plot(x, y1)

pl.show()
```

(2) 커스터마이징하기

```python
from pylab import *
import numpy as np

# generate uniformly distributed
# 256 points from -pi to pi, inclusive
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# these are vectorised versions
# of math.cos, and math.sin in built-in Python maths
# compute cos for every x
y = np.cos(x)

# compute sin for every x
y1 = np.sin(x)

# plot cos
plot(x,y)

# plot sin
plot(x,y1)

# define plot title
title("Functions $\sin$ and $\cos$")

# set x limit
xlim(-3.0, 3.0)
# set y limit
ylim(-1.0, 1.0)

# format ticks at specific values
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

show()
```

- title()로 제목을 붙일 수 있다. 
- xlim()과 ylim()으로 x축과 y축의 범위를 정할 수 있다. 
- xticks()와 yticks()로 ticker에 표시되는 값을 변경할 수 있다. 
- LaTex문법인 $\sin$, or $-\pi$를 이용하여 그리스문자를 표시한다(나중에 다시 다룬다).

## 3.3 Defining axis lengths and limits

(1) axis의 길이 변경하기

```python
axis()
l = (0.0, 1.0, 0.0, 1.0)
axis(l)
```

- 실행해보면 (0.0, 1.0, 0.0, 1.0)이 출력됨을 알 수 있다. 
- 이것은 각각 xmin, xmax, ymin, ymax값을 표시한다. 

```python
axis()
l = [-1, 1, -10, 10]
axis(l)
```

- 값을 [-1, 1, -10, 10]으로 바꿔서 실행해보면 x,y축의 범위가 바뀌었음을 볼 수 있다. 

- axis()를 쓰지 않으면 matplotlib은 자동적으로 하나의 플롯에서 모든 데이터 포인트가 보일 수 있는 최소값을 사용한다. 만약 axis()의 limit을 데이터셋의 최대값보다 작게 설정할 경우 모든 데이터 포인트가 보이지 않게 될 수 있다. 
- 혼란을 피하기 위해서는  autoscale()을 사용해 데이터에 맞는 최적의 축 사이즈를 계산하도록 할 수 있다. (matplotlib.pyplo.autoscale())

(2) axis를 추가하기

여러 개의 차트를 결합하여 같은 데이터와 관련된 다른 뷰를 보여줄 때 유용하게 사용할 수 있다. 
- 같은 figure에 새로운 axis를 추가하려면 matplotlib.pyploy.axes()를 사용한다. 
- rect와 axisbg : 축의 배경색을 변경할 수 있다.
- sharex/sharey : 추가된 축의 속성을 지정할 수 있다. 
- polar : polar axes를 사용할 수 있다. 

- 현재 figure에 line을 추가하려면 matplotlib.pyploy.axhline() 또는 matplotlib.pyplo.axvline()을 사용한다. 
- axhline에는 y position, xmin, xmax를, axvline() 에는 x position, ymin, ymax를 지정할 수 있다.  

```python
axis()
l = [-1, 1, -10, 10]
axis(l)

axhline()
axvline()
axhline(4)

show()
```

파라미터값을 주지 않으면 y = 0 (axhline())과 x = 0 (axvline())이 대입된다. 

추가적인 기능들 : 
- matplotlib.pyplot.axhspan()과 matplotlib.pyplot.axspan()을 이용하여 horizontal span(rectangle)을 추가할 수 있다. 
- axhspan은 ymin과 ymax를 파라미터로 받고, axspan은 xmin과 xmax를 받는다. 

그리드 보이기 : 
- matplotlib.pyplot.grid()는 그리드를 보이거나 안 보이게 할 수 있다. 
- 파라미터로 그리드의 tick type을 정하는 which와 x,y,both 중 무엇을 보여줄지 정하는 axis를 받는다. 

축을 결정하는 것 : 
- matplotlib.pyplot.axis()의 부모 클래스는 matplotlib.axes.Axes이다. 
- 하나의 축은 matplotlib.axis.Axis이며, x축은 matplotlib.axis.XAxis이고, y축은 matplotlib.axis.YAxis이다. 
- 이러한 기능을 이 책에서 다루지는 않지만, 고급 axis control 이 필요할 경우를 위해 알아두어야 한다.

##3.5 Setting ticks, labels, and grids

* matplotlib에서 figure()는 user interface window를 생성하며, 2개 이상의 subplots를 가질 수 있다. 
Subplots는 그리드에 plots를 생성한다. 
보다 복잡한 컨트롤을 하려면 matplotlib.axes.Axes의  axes 인스턴스를 사용할 필요가 있다. 

* Ticks는 figure의 일부로서, Major ticks와 minor ticks는 서로 독립적이다. 
matplotlib.pyplot.locator_params() 을 사용하여 tick 위치를 조절할 수 있다. 

```python
from pylab import * 
# get current axis 
ax = gca()

# set view to tight, and maximum number of tick intervals to 10 
ax.locator_params(tight=True, nbins = 10) 

# generate 100 normal distribution values 
ax.plot(np.random.normal(10, .1, 100)) 

show()
```

* tick formatters는 숫자값들이 보여지는 방법을 결정한다. 
* matplotlib.ticker.FormatStrFormatter를 사용한다. 
* Helper function으로 matplotlib.dates.date2num(), matplotlib.dates.num2date(), 
그리고 matplotlib.dates.drange()을 사용하여 날짜를 다른 형식으로 변환할 수 있다. 

```python
from pylab import *
import matplotlib as mpl
import datetime

fig = figure()

# get current axis
ax = gca()

# set some daterange
start = datetime.datetime(2013, 01, 01)
stop = datetime.datetime(2013, 12, 31)
delta = datetime.timedelta(days = 1)

# convert dates for matplotlib
dates = mpl.dates.drange(start, stop, delta)

# generate some random values
values = np.random.rand(len(dates))

ax = gca()

# create plot with dates
ax.plot_date(dates, values, linestyle='-', marker='')

# specify formater
date_format = mpl.dates.DateFormatter('%Y-%m-%d')

# apply formater
ax.xaxis.set_major_formatter(date_format)

# autoformat date labels
# rotates labels by 30 degrees by default
# use rotate param to specify different rotation degree 
# use bottom param to give more room to date labels
fig.autofmt_xdate()

show()
```

* 데이터 레이블의 포맷을 변환한다
* 레이블을 30도 회전한다
* rotate 파라미터로 회전각도를 지정한다
* bottom 파라미터로 날짜 레이블에 여백을 준다. 

##3.6 Adding legends and annotations

범례와 각주는 data plot을 명확하고 맥락에 맞게 설명해줌으로써 독자의 머리 속에 좀더 쉬운 멘탈 모델을 형성할 수 있도록 한다. 

```python
from matplotlib.pyplot import *

# generate different normal distributions
x1 = np.random.normal(30, 3, 100)
x2 = np.random.normal(20, 2, 100)
x3 = np.random.normal(10, 3, 100)

# plot them
plot(x1, label='plot')
plot(x2, label='2nd plot')
plot(x3, label='last plot')

# generate a legend box
legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
       ncol=3, mode="expand", borderaxespad=0.)

# annotate an important value
annotate("Important value", (55,20), xycoords='data',
         xytext=(5, 38), 
         arrowprops=dict(arrowstyle='->')) 
show()
```

##3.7 Moving spines to the center

* spine 을 중심으로 옮기는 방법
* spine은 데이터 영역의 경계를 정의한다. 기본값은 축의 가장자리에 있기 때문에 플롯 주위에 상자를 볼 수 있다. 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 500, endpoint=True) 
y = np.sin(x)

plt.plot(x, y)

ax = plt.gca()

# hide two spines 
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# move bottom and left spine to 0,0
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

# move ticks positions
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.show()
```

set_smart_bounds(True)를 이용하여 spine을 데이터가 끝나는 지점으로 제한할 수 있다. 

##3.8 Making histograms

* 히스토그램은 세로형의 사각형으로 bin이라 부르는 특정 간격 내에서 데이터 포인트의 빈도를 표시한다. 
* 히스토그램은 데이터의 상대빈도를 표시하므로 총면적은 1이 된다. 
* 여기서는 2D 히스토그램을 다루고, 5장에서 3D  히스토그램을 다룰 것이다. 

matplotlib.pyploy.hist()으로 히스토그램을 그린다. 유용한 파라미터는 다음과 같다. 
* bins -  기본값은 10이다 
* range  기본값은  None이고 아웃라이어는 무시된다
* normed – 기본값은 False이고, True로 할 경우 값이 정규화된다
* histtype – 기본적인 막대형 히스토그램이다. barstacked, step, stepfilled의 옵션을 줄 수 있다. 
* align – 기본값은 mid이다. left나 right  옵션을 줄 수 있다
* color – 단일색상이나 연속색상을 사용할 수 있다
* orientation – 기본값은 vertical이고, horizontal옵션을 줄 수 있다

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 10000)

ax = plt.gca()

# the histogram of the data
ax.hist(x, bins=35, color='r')

ax.set_xlabel('Values')
ax.set_ylabel('Frequency')

ax.set_title(r'$\mathrm{Histogram:}\ \mu=%d,\ \sigma=%d$' % (mu, sigma))

plt.show()
```

* bin을 35로 주었으며, 정규화를 True(1)로 주었고, 색상은 red로 주었다
* LaTeX를 이용하여 수학기호를 혼합하여 썼다. 

##3.9 Making bar charts with error bars

error bar는 표준편차, 표준오차, 또는 95% 신뢰구간을 보여줄 수 있다. 기준은 없으므로 error bar가 어떤 값을 나타내는지 언급해주어야 한다. 

반드시 사용해야하는 파라미터는 left와 height이다. 다른 옵션은 다음과 같다. 
* width –기본값은 0.8
* botton – 기본값은 None
* edgecolor – bar edge의 색상
* ecolor – error bar의 색상
* linewidth – 기본값은 None과 0(표시되지 않음)
* orientation – vertical과 horizontal을 쓸 수 있다
* xerr and yerr – error bar를 생성

```python
import numpy as np
import matplotlib.pyplot as plt

# generate measures from gaussian distribution 
x = np.arange(0, 10, 1)

# values computed from "measured"
y = np.log(x)

# add some error samples from standard normal distribution 
xe = 0.1 * np.abs(np.random.randn(len(y)))

# draw and show errorbar
plt.bar(x, y, yerr=xe, width=0.4, align='center', ecolor='r', color='cyan', label='experiment #1')

# give some explainations
plt.xlabel('# measurement')
plt.ylabel('Measured values')
plt.title('Measurements')
plt.legend(loc='upper left')

plt.show()
```

여기에 더할 수 있는 옵션은 p.101 을 참조( / \ | - + x o 0 . * ) 

```python
Image(filename='PyVizImg/errorbar01.png')
```

비대칭의 error bar를 그리려면 xerr 또는 yerr를 2개의 요소를 가진 list로 만들면 된다. 
* 첫번째 리스트는 negative error를, 두번째 리스트는 positive error를 갖도록 한다. 

##3.10 Making pie charts count

* Pie chart는 데이터셋의 총합이 100%가 되도록 그려야 한다. 
* Pie chart는 비율을 표시하며, 간편하고 미학적으로 보기 좋을 수 있지만 비교가 어렵다는 비판도 받는다. 
* Pie chart는 특정 각도나 색상을 쓸 때 잘 작동하지 않으며, 우리의 지각을 왜곡시키고 우리의 결정에 영향을 미칠 수 있다. 

우선 exploded pie chart를 그려보자. 

```python
import matplotlib.pyplot as plt

# make a square figure and axes
# pie chart looks best in square figures
# otherwise it looks like ellipses
plt.figure(1, figsize=(8, 8))
ax = plt.axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = 'Spring', 'Summer', 'Autumn', 'Winter'
values = [15, 16, 16, 28]
explode =[0.1, 0.1, 0.1, 0.1]

# make a pie 
plt.pie(values, explode=explode, labels=labels,
    autopct='%1.1f%%', startangle=67, shadow=False)

plt.title('Rainy days by season')

plt.show()
```

* autopct 파라미터는 레이블을 원호 내부에 그리는 데 사용한다. 
* boolean shadow 파라미터는 파이차트에 그림자 효과를 더한다.
* Startangle은 x축(0도)으로부터 반시계방향으로 그리고, y축에서부터 그리려면 90도로 지정하면 된다. 

##3.11 Plotting with filled areas

커브 아래 또는 두 개의 커브 사이 영역을 채워보자. 

```python
from matplotlib.pyplot import figure, show, gca
import numpy as np

# generate x values
x = np.arange(0.0, 2, 0.01)

# two different signals are measured
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)

fig = figure()
ax = gca()

# plot and 
# fill between y1 and y2 where a logical condition is met
ax.plot(x, y1, x, y2, color='black')

ax.fill_between(x, y1, y2, where=y2>=y1, facecolor='darkblue', interpolate=True)
ax.fill_between(x, y1, y2, where=y2<=y1, facecolor='deeppink', interpolate=True)

ax.set_title('filled between')

show()
```

* hatch  (패턴으로 채우기)와 line 옵션(linewidth, linestyle)을 더할 수 있다. 
* fill_btetweenx()는 수평 커브 사이에서 면적을 채워준다. 
* 좀더 일반적인 함수인 fill()은 어떤 다각형이든 색상이나 패턴으로 채워준다. 

##3.12 Drawing scatter plots with colored markers

* 2개의 변수 사이의 상관을 보여주고 싶을 때  scatter plot을 사용한다. 
* 이것은 scatter plot matrix와 같은 다차원 데이터의 고급 시각화를 위한 시작으로 유용하다. 

```python
import matplotlib.pyplot as plt
import numpy as np

# generate x values
x = np.random.randn(1000)

# random measurements, no correlation
y1 = np.random.randn(len(x))

# strong correlation
y2 = 1.2 + np.exp(x)

ax1 = plt.subplot(121)
plt.scatter(x, y1, color='indigo', alpha=0.3, edgecolors='white', label='no correl')
plt.xlabel('no correlation')
plt.grid(True)
plt.legend()

ax2 = plt.subplot(122, sharey=ax1, sharex=ax1)
plt.scatter(x, y2, color='green', alpha=0.3, edgecolors='grey', label='correl')
plt.xlabel('strong correlation')
plt.grid(True)
plt.legend()

plt.show()
```

* scatter plot은 두 변수간의 잠재적 상관을 확인하기 위해 흔히 사용되며, 회귀 함수의 피팅 작업 전에 그려진다. 
* 이것은 상관관계를 시각적으로 잘 보여주는데, 특히 비선형 관계에 유용하다.  Matplotlib은  scatter() 함수를 제공한다.
