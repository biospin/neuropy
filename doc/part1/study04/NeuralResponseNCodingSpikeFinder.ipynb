##2. 활동전위 (Action potential)

### 신경세포는 기본적으로 절연체인 세포막으로 쌓여있으며, 세포 내외에는 전해질 용액인 나트륨($Na^+$)과 칼륨($K^+$) 등의 이온화 된 전해질 금속을 포함하고 있다. 세포는 이러한 이온물질을 이용해 전위를 발생시키기 위해 다름과 같은 구조물을 포함하고있음
- $Na^+$/$K^+$ 펌프 - 항상 $Na^+$을 안쪽에서 바깥쪽으로, $K^+$을 바깥쪽에서 안쪽으로 수송. 따라서 항상 세포 안쪽은 $K^+$, 세포 바깥쪽은 $Na^+$ 농도가 높음
- $K^+$ 채널 - 채널이 열리게 되면 농도구배에 의해 $K^+$가 세포 안쪽에서 바깥쪽을 향하여 이동함: 외향 K전류
 - 정지상태에서 세포 안쪽은 바깥쪽에 비해 -50 ~ 100mV 정도 전기적으로 음(-)전하를 띔: 분극(polarization)
 - 안정적으로 분극되어있는 상태의 전위차를 세포의 안정막전위(resting membrane potential, resting potential, RMP)라고 함
 - 신경세포의 흥분성 변화는 세포의 안정막전위의 변화로 나타남
    
    
- $Na^+$ 채널 - 채널이 열리게 되면 농도구배에 의해 $Na^+$가 세포 바깥쪽에서 안쪽을 향하여 이동함: 내향 Na전류
 - 급격한 양(+)전하화: 탈분극(depolarization), 약 1 ms 정도 지속됨
 - 0 전위를 넘어감으로, 초과전위(역전전위, overshoot) 라고 함

### 탈분극한 상태에서 안정막 전위로 돌아오는 과정을 재분극(repolarization)이라고 함, 이 과정에 안정막 전위보다 더 낮은 전위가 일시적으로 나타나는 현상을 과분극(hyperpolarization) 또는 Undershoot 라고 함

### 세포막 전위가 역치전위(threshold)인 20~40 mV 이상의 양전하까지 도달하게되면 일시적으로 전하의존성 $Na^+$ 채널이 일시적으로 열리게 되어 100~120 mV 의 전위변화가 약 0.5 ~ 몇 ms 의 매우 짧은시간동안 나타나며 이를 활동전위(action potential)이라 함

![image](img/AP.jpg)

##3. 흥분의 전도
### 신경섬유의 종말 부분은 다른세포의 수상돌기와 연접에 의해 결합
- 신경섬유에서 전기전도는 양쪽 방향으로 전해지나, 연접에서는 반드시 한쪽의 세포로부터 화학물질(아세틸콜린 등의 신경전달물질)이 분비되고 연접 후막(postsynaptic membrane)의 수용기가 그것을 받아 다시 한쪽 방향으로만 전달하여 전기신호로 변환 함
- ** 흥분성 신경전달물질 **: 아세틸콜린(acetylcholine), 노르아드레날린(noradrenalin)
- ** 억제성 신경전달물질 **: GABA ($\gamma$-aminobutylic acid)
- 신경전달물질에 의한 전위는 활동전위보다 전압은 낮으나 일정시간 지속되는(수 ms) 전위변화를 갖는데 이를 연접후전위(postsynaptic potential, PSP)라고 하며 ** 흥분성 연접후전위 (excitatory postsynaptic potential, EPSP) **와 ** 억제성 연점후전위 (inhibitory postsynaptic potential, IPSP) ** 로 구분됨

##4. 실무율( All or none law)
### 자극강도가 역치 이하이면 활동전위는 전혀 생기지 않고, 문턱 이상이면 일정 크기의 활동전위가 생기는 현상
- 단일 신경세포의 활동전위는 자극강도에 비례하여 커지지 않음: 하지만 신경세포 다발의 활동전위는 자극강도에 비례하여 커짐

# 뇌파의 측정

### 뇌파측정으로 얻어지는 신호는 주로 대뇌피질의 추체세포(Pyramidal cell)의 심부 혹은 표층의 수상돌기(dendrites) 에서의 활동전위 및 연접후전위(EPSP, IPSP) 등의 종합이 집적되어 비교적 완만한 파형이 형성된 것을 두피(scalp)위에서 기록한 것, 시상(Thalamus) 이라는 뇌 구조는 뇌파의 리듬을 얻는 길잡이 (pacemaker)로 생각됨

### 뇌파를 구성하는 파의 대부분은 전위변동이 sine curve거나 이와 유사한 sinusoid curve로 파형, 주파수가 거의 같은 상향파가 연속되어 나타며, 뇌파소견을 종합적으로 판정하기 위해 기술하는 구성요소는 아래와 같은 것들이 있음

##1. 구성요소

###1) 활동 (Activity)
- 연속된 전위변동의 출현을 의미하며 보통 전기활성(electrical activity)와 같은 의미로 쓰인다.

###2) 파 (Wave)
- 뇌파에서 어떤 단일 또는 일과성의 전위변동

###3) 율동 (Rhythm)
- 일정한 주기(period)와 파형을 가지고 반복되는 파로 이루어진 뇌파활동, 진폭(amplitude)은 일정하지 않아도 됨

##2. 뇌파소견을 표현하는 지표

###1) 주파수(frequency)와 주기(period)
![Frequency](img/Frequency.png)

- 주파수는 1초동안 반복하여 출현하는 율동적인 파의 횟수 (Hz)

![Amplitude](img/Period.gif)
- 주기는 1개의 파의 산과 산 또는 계곡과 계곡을 연결하여 그 간격에 이르는 시간적 길이를 의미(지속시간, duration)

    - $\delta$ wave : 0.5 ~  3 Hz 
    - $\theta$ wave :   4 ~  7 Hz - 느린파 ~
    - $\alpha$ wave :   8 ~ 13 Hz - 기본파
    - $\beta$ wave  :  14 ~ 30 Hz - ~ 빠른파
    - $\gamma$ wave :       30 Hz 이상
    
###2) 진폭(amplitude) 및 전위(potential)

- 진폭은 파의 계곡과 계곡을 연결한 선과 산의 정점(peak)으로부터의 기선에 수직선을 내려 긋고 그것과 만나는 점까지의 거리(h)를 말함
- 전압은 정해진 교정전압을 가했을 때, 교정곡선의 진폭과 비교해서 $\micro$V 로 나타냄, 교정곡선의 진폭은 50 $\micro$V 를 5 mm 가 되도록 조정 (10 $\micro$V = 1 mm) 하므로 측정치의 진폭과 교정곡선의 진폭과의 비를 얻으면 전압을 얻을 수 있음

###3) 위상(phase)  및 위상관계(phase relation)
- 뇌파로 기록된 파형의 다른부위에서의 위치 및 시간관계

###4) 분포(distribution)
- 어떤 이상파 또는 뇌파활동이 두피위에서 측정되는 범위

###5) 반복성(repetition)
- 율동성(rhythmical): 각각의 파가 규칙적으로 출현하며 sine 파 형태로 나타날때 
- 비율동성(arrhythmical): 각각 주파수가 다른 파가 계속된 상태로 다형성(polymorphic)한 형태로 나타날때

###6) 반응성(reactivity)
- 환경의 변화나 외부의 자극에 대응하는 뇌파의 가변적 변화를 의미함

###7) 지속성(persistence)
- 뇌파기록 중에 일정한 파형이나 패턴이 얼마나 자주 출현하는가, 어떤 종류의 파가 어느 정도의 양으로 출현하고 있는가를 나타내며 출현양(quantity)이라고 표현하기도 함

###8) 변화성(variability)
- 특정한 파가 지속적으로 나타나는지, 불규칙한 시간간격으로 반복되는지, 일정하지 않은 간격으로 반복하나 돌발성으로 나타나는지

###9) 출현양식
- 어떤 파가 어느 정도의 양으로 출현하고 있는가를 나타내는 것

 ####(1) 돌발성(paroxysmal)
 - 배경파에서 이상파가 일과성으로 갑자기 나타났다가 없어지는 경우

 ####(2) 비돌발성(non-paroxysmal) 또는 지속성(persistent, continuous)
 - 어떤 파가 도중에 끊어지지 않고 계속해서 나오고 있는 경우

 ####(3) 간헐성(intermittent)
 - 돌발성과 지속성의 중간정도

###10) 시간적 관계
- 어떤 부위가 다른 부위보다 시잔적으로 빠르게 나타나면서 점차 다른 부위로 파급되는 경우, (시간적으로 앞에 나타나는 부분에 원발성 초점이 있음)

### 조금 다르지만 거의 비슷한 곳에 그래프가 그려짐을 볼 수 있다. 위 방법 중 한가지 방법을 이용해 함수를 만들자

```python
def ap_finder(time,voltage, threshold): # 데이터와 threshold 를 입력한다.
 
    aptimes = []
       
    # 혹시 모르니까 입력되는 데이터에 문제가 있는지를 체크한다.
    if (len(voltage) != len(time)):
        print "Can't run - the vectors aren't the same length!"
        return aptimes
    
    crossThreshold = np.logical_and(voltage[:-1] >= threshold,
                                    voltage[1:] < threshold)
    
    aptimes = time[crossThreshold]
    
    return aptimes
```

### 이제 원래 그래프에 활성전위를 표시하여 그래프를 그려보자

```python
def plot_spikes(time,voltage,APTimes,titlestr):
    plot(time,voltage,hold=True)  # 먼저 plot 을 그림
    p = np.ones(len(APTimes))     # 체크된 활동전위와 동일한 수의 모든 값이 1로 된 임의의 np.array 를 만듦
    p = p*max(voltage)+25         # p 값을 voltage 의 최대값에 25를 더한값으로 지정 (plot 할 위치 지정)
    plot(APTimes,p,'|r',hold=True) # |r 은 붉은 막대기 모양, hold=True 는 덮어쓰기
    xlabel('Time (s)')
    ylabel('Voltage (uV)')
    title(titlestr)
```

###이제 활성전위로 표시된 부분만 추출해서 겹쳐지게 그래프를 그려봅시다

```python
def plot_waveforms(time,voltage,APTimes,titlestr):
    
    sample_rate = float(len(time)-1)/(1000*(time[-1]-time[0]))
    
    for i in range(0, len(APTimes)):  
        idx_spike = min(find(abs(time - APTimes[i])<0.00001))
        #to allow 0.01ms jittering
        #I use min() here in case it returns several indices
               
        # find the start and end index for plotting
        idx_start = idx_spike - int(3*sample_rate)
        idx_end   = idx_spike + int(3*sample_rate)

        if (idx_start>=0) & (idx_end < len(time)-1):
            # plot the waveform for the i-th spike
            xx = np.linspace(-3,3,sample_rate*6)
            yy = voltage[range(idx_start, idx_end)]
            plot(xx, yy, 'b',hold=True,)
    
    xlabel('Time (ms)')
    ylabel('Voltage (uV)')  
    title(titlestr)
```
