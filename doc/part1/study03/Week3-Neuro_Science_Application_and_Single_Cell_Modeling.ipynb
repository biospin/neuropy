###전통적 모델링

####수학

뉴턴의 냉각법칙: 시간, $t$의 변화에 따른 어떤 물체의 온도, $T_o$의 변화량은 그 물체와 주변 온도, $T_s$의 차이에 비례($k$)한다. 이를 수학적으로 아래와 같이 정의(모델링) 할 수 있다. 단, 대류로 인한 온도 변화는 없다고 가정한다.

\begin{equation*}
\frac{dT_o}{dt} = k(T_o - T_s)
\;\;\;
\mathit{(where, T_o: 어느 물체의 온도, T_s: 주변 온도, k: 비례 상수, t: 시간)}
\end{equation*} 

이 정의 해를 구하는 과정은 아래와 같다.

\begin{equation*}
\frac{dT_o}{(T_o - T_s)} = kdt
\end{equation*}

\begin{equation*}
ln(T_o - T_s) = \int kdt
\end{equation*}

\begin{equation*}
ln(T_o - T_s) = kt + C
\end{equation*}

\begin{equation*}
T_o - T_s = e^{kt + C}
\end{equation*}

\begin{equation*}
T_o = \bar{C}e^{kt}  + T_s
\;\;\;
\mathit{(where, \bar{C} = e^C)}
\end{equation*}

또는,  $T_o$에 대한 함수로 정리하면

\begin{equation*}
T_o(t) = \bar{C}e^{kt}  + T_s { }
\end{equation*}

리비의 탄소 연대 추정법: $\frac{dC}{dt} = kC$

멜서스의 인구 법칙: $\frac{dP}{dt} = kP$

약동학: Two compartment (tank) model $Y(t) = Y_{upper}(T) + Y_{lower}(T)$

###Simulation 1

[Tutorial 1: modeling a single compartment](http://pub.ist.ac.at/~jguzman/doc/neuron/tutorial/Tutorial2.html) 참조. 일부 코드는 실행을 위해 수정하였음.

```python
# import neuron and nrn modules
import neuron, nrn
```

```python
# Creating and accession sections
soma = nrn.Section()

# soma has 3 segments
# each segment has length of 30 micro-meter, 30 micro-meter of diameter
soma.L = 30
soma.nseg = 3
soma.diam = 30
```

```python
# check soma's capacitance (mF / cm^2)
print(soma.cm)
```

```python
# check soma's Cytoplasmic resistivity (Ohms-cm)
soma.Ra
```

```python
# check the different segments in a section
# this will retern corresponding values (location, diameter and capacitance)
# of the different segments
for nseg in soma:
    print(nseg.x, nseg.diam, nseg.cm)
```

```python
# access an individual segment
soma_seg = soma(0.5)
print(soma_seg.diam)
```

```python
# handle the segment properties with list
soma_nseg = [] # create an empty list of segments
for nseg in soma:
    soma_nseg.append(nseg.x)

print(soma_nseg)
```

```python
# access with list
print(soma(soma_nseg[1]).diam) # same as upper code soma_seg = soma(0.5); print(soma_seg.diam)
```

```python
# we can change properties, diameter
soma(soma_nseg[0]).diam = 10
soma(soma_nseg[2]).diam = 10
```

```python
# check the different segments in a section
# this will retern corresponding values (location, diameter and capacitance)
# of the different segments
for nseg in soma:
    print(nseg.x, nseg.diam, nseg.cm)
```

```python
# not like in hoc, there is not such thing like "access soma" for accessing a section.
# when we create the soma section, it became automatically the currently accessed by default.
# To know which is the currently accessed section we use the cas() method.
# cas() means currently accessed section
print(nrn.cas().name())
```

```python
# you can get soma name
print(soma.name())
```

```python
# adding membrane properties
# here, we put Hodgkin-Huxley conductances
soma.insert("hh")
```

The Hodgkin-Huxley channels add the following new properties to the section
* **gnabar_hh**: The maximum specific sodium(Na) channel conductance [Default value = 0.120 S/$cm^2$]
* **gkbar_hh**: The maximum specific potassium(K) channel conductance [Default value = 0.036 S/$cm^2$]
* **gl_hh**: The maximum specific leakage conductance [Default value = 0.0003 S/$cm^2$]
* **ena**: The reversal potential for the sodium channel [Default value = 50 mV]
* **ek**: The reversal potential for the potassium channel [Default value = -77 mV]
* **el_hh**: The reversal potential for the leakage channel [Default value = -54.3 mV]

```python
# check the 2nd segment's voltage of the soma
print(soma(0.5).v)
```

```python
# check the reversal potential for the sodium channel
# this should be default value
print([soma(nseg.x).ena for nseg in soma])
```

```python
# the potassium reversalpotential in the second (0.5) segment
print(soma(0.5).ek)
```

```python
# same result with above code block
print(soma(soma_nseg[1]).ek)
```

```python
# Point processes: simulation
stimulator = neuron.h.IClamp(0.5, soma)
```

The built-in point process 'IClamp' has its own properties:
* **delay** is the delay until the onset of the stimulus (in msec)
* **dur** is the duration of the stimulus (in msec)
* **amp** is the amplitude of the stimulus (in nA)

```python
stimulator.delay = 0.1
stimulator.dur = 0.8
stimulator.amp = 1.2
```

```python
# Running the simulation

# to do simulation calling the hoc interpreter
from neuron import h
```

```python
# check initial state
print(soma(0.5).v)
```

```python
# call hoc interpreter to run "run()" command
h.load_file("stdrun.hoc")              
h.init()
h.run()
```

```python
# check changed state
print(soma(0.5).v)
```

```python
print(h.tstop)
h.tstop = 1.5
h.run()
print(soma(0.5).v)
```
