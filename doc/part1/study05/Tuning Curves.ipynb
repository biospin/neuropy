```python
def bin_spikes(trials, spk_times, time_bin = 0.1):
    start_bin = time_bin
    stop_bin = time_bin
    
    trial_times = trials[:,1]
    spikes_per_trial = np.zeros(len(trial_times))
    
    for t in np.arange(0,len(trial_times)):
        # http://docs.scipy.org/doc/numpy/reference/generated/numpy.logical_and.html
        spikes_per_trial[t] = np.count_nonzero(
np.logical_and(spk_times > trial_times[t] - start_bin, 
               spk_times < trial_times[t] + stop_bin))
    
    directions = np.unique(trials[:,0])
    rates = np.zeros(len(directions))
    
    for dirs in range(len(directions)):
        d  = directions[dirs]
        indices = plt.find(trials[:,0] == d)
        rates[dirs] = sum(spikes_per_trial[indices])/len(indices) 
    rates = rates/(start_bin + stop_bin)  
    dir_rates = np.column_stack((directions,rates))
    return dir_rates
```

```python
dir_rates = bin_spikes(trials, spk_times, 0.1)
```

```python
from scipy import stats

def von_mises_fitfunc(x, A, kappa, l, s):
    return A*stats.vonmises.pdf(x, kappa, loc=l, scale=s)
```

```python
x = dir_rates[:,0]
y = dir_rates[:,1]
peak_guess = np.deg2rad(x[np.argmax(y)])
guesses = [np.amax(y),4.,peak_guess,1]
p, cov = optimize.curve_fit(von_mises_fitfunc, np.deg2rad(x), y, guesses)

curve_xs = np.arange(0,360)

fit_ys = von_mises_fitfunc(np.deg2rad(curve_xs),p[0],p[1],p[2],p[3])
```
